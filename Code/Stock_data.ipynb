### Concating the dataframe to eventually make a final dataframe

```python
df_final = pd.concat([cfg_data,ms_data,cme_data,jpm_data,gs_data,pypl_data,td_data,bmo_data,bns_data,usb_data,axp_data,spx_data,dow_data,nasdaq_data,vix_data]).drop_duplicates()
df_final.head()
```

```python
# Sorting values of dates in decending order
df_sort = df_final.sort_values(['Date'], ascending=[False])
# to do reset index
df_stock = df_sort.reset_index()
df_stock.head(10)
```

```python
df_stock = df_stock.rename(columns={"Adj Close": 'Adj_Close'})

df_stock.head()
```

```python
df_stock = df_stock[["Date", "Symbols", "High", "Low", "Open", "Close", "Volume", "Adj_Close", "Company_name"]].copy()
df_stock.head(10)
```
